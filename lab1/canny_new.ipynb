```python
import cv2
import os
import requests
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import keras_core as keras
```

```python
I_o = keras.utils.load_img("kruwsko.png")
plt.imshow(I_o)
```

```python
I_o = keras.utils.img_to_array(I_o)
I=I_o/255
```

```python
# RGB_to_GRAY = keras.layers.Conv2D(
#   filters=32,
#   kernel_size=(1,1,1),
#     strides=(1,1,3),
#     padding="valid",
#   )
```

```python
I_gray = np.zeros((I.shape[0],I.shape[1]))
for y in range(I.shape[0]):
  for x in range(I.shape[1]):
    # NTSC fromula
    I_gray[y][x] = 0.299 * I[y][x][0] + 0.587 * I[y][x][1] + 0.114* I[y][x][2]
```

```python
plt.imshow(I_gray,"gray")
```

```python
max_pool_2d = keras.layers.MaxPool2D(
  pool_size=(4,4), strides=(4,4), padding="valid"
)
avg_pool_2d = keras.layers.AveragePooling2D(
    pool_size=(4, 4), strides=(4, 4), padding="valid"
)
resize_to_4D = I_gray.reshape([1,I.shape[0],I.shape[1],1])
I_max = max_pool_2d(resize_to_4D)
print(I_max)
plt.imshow(I_max.reshape([I_max.shape[1], I_max.shape[2]]), "gray")
plt.show()
I_avg = avg_pool_2d(resize_to_4D)
plt.imshow(I_avg, "gray")
plt.show()
```
